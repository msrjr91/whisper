```python
# import/install dependencies
from faker import Faker
# from random import randint
import csv
# import datetime
```

```python
# initialize faker function
fake = Faker()
```

```python
# example:
print(fake.user_name())
```

```python
# all model fields
user_fields = ['first_name', 'last_name', 'email', 'password', 'username']
```

```python
# generate fake user profile information required for registration process for an account
def make_users(num_records, headers):
    with open("user_data.csv", "wt") as csvFile:
        writer = csv.DictWriter(csvFile, fieldnames=headers)
        writer.writeheader()
        for i in range(num_records):
            writer.writerow({
                "first_name": fake.first_name(),
                "last_name": fake.last_name(),
                "email": fake.email(),
                "password": fake.password(length=20, special_chars=True, upper_case=True),
                "username": fake.user_name()
            })
            
```

```python
# MAKE USERS
make_users(100,user_fields)
```
